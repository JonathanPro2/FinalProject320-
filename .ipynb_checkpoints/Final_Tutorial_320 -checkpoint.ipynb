## Data Scraping

The dataset that we are using is located at https://covid.ourworldindata.org/data/owid-covid-data.csv

It updates daily, so we are using the dataset as it was downloaded on May 03, 2021.

The dataset's README at https://github.com/owid/covid-19-data/tree/master/public/data shows that it contains daily records for cases, deaths, and health statistics for every country in the world. 

```python
data = pd.read_csv('owid-covid-data.csv') 
data.head() 
```

## Data Cleaning

Besides daily information for each country, the data also contains daily information for each continent as well as the entire world. We remove the rows that contain a non-country value as its location. This is represented in the data as a iso_code starting with OWID.

```python
# deleting references to world data (wrl)
data = data[data.iso_code != 'OWID_WRL']

# # deleting references to Africa (afr)
data = data[data.iso_code != 'OWID_AFR']

# # deleting references to Asia (asi)
data = data[data.iso_code != 'OWID_ASI']

# # deleting references to Europe and EU (EUR and EUN)
data = data[data.iso_code != 'OWID_EUR']
data = data[data.iso_code != 'OWID_EUN']

# # deleting references to North America (nam)
data = data[data.iso_code != 'OWID_NAM']

# # deleting references to Oceania (oce)
data = data[data.iso_code != 'OWID_OCE']

# # deleting references to South America (sam)  
data = data[data.iso_code != 'OWID_SAM']
```

Additionally, for initial data exploration, we will only be looking at cases and deaths for the top 50 most populous countries on May 3rd 2021. We drop the unneeded columns.

```python
# Sorting the data by population, we will only look at the top 50 most populous countries
data = data.sort_values(by=['population'] , ascending = False) 

# remove data columns that will not be used
data = data.drop(columns=['reproduction_rate', 'icu_patients', 'icu_patients_per_million', 'hosp_patients', 
                         'hosp_patients_per_million', 'weekly_icu_admissions', 'weekly_icu_admissions_per_million', 
                         'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million', 'new_tests', 'total_tests', 
                         'total_tests_per_thousand', 'new_tests_per_thousand', 'new_tests_smoothed', 
                         'new_tests_smoothed_per_thousand', 'positive_rate', 'tests_per_case', 'tests_units', 
                         'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations', 
                         'new_vaccinations_smoothed', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
                         'people_fully_vaccinated_per_hundred', 'new_vaccinations_smoothed_per_million', 
                         'stringency_index', 'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 
                         'gdp_per_capita', 'extreme_poverty', 'cardiovasc_death_rate', 'diabetes_prevalence', 
                         'female_smokers', 'male_smokers', 'hospital_beds_per_thousand', 
                         'life_expectancy', 'human_development_index', 'handwashing_facilities'])

data = data.drop(columns=['new_cases_smoothed', 'new_deaths_smoothed', 'total_cases_per_million', 'new_cases_per_million', 
                         'new_cases_smoothed_per_million', 'total_deaths_per_million', 'new_deaths_per_million', 
                         'new_deaths_smoothed_per_million'])


# getting the data for the latest date available (5/3/2021)
top_50 = data[data.date == '2021-05-03']
top_50 = top_50.head(50)

# create new columns for the averages of new deaths and new cases
# iter through rows of top_50 using the location to get the mean of all new cases/new deaths from that location
avg_new_cases = []
for index, row in top_50.iterrows():
    avg_new_cases.append(data[data.location == row['location']]['new_cases'].mean())
    
avg_new_deaths = []
for index, row in top_50.iterrows():
    avg_new_deaths.append(data[data.location == row['location']]['new_deaths'].mean())
    
# add new columns to top_50
top_50['avg_new_cases'] = avg_new_cases
top_50['avg_new_deaths'] = avg_new_deaths

# remove new cases and new deaths columns as they are only for may 3
top_50= top_50.drop(columns=['new_deaths', 'new_cases'])
# remove date because data spans more than just this date
top_50 = top_50.drop(columns=['date'])


top_50
```

## Data Exploration and Visualization

Data exploration is an important step because it provides an idea on relationships that might exist within the data. This helps guide our data analysis with hypothesis testing and machine learning.

The data contains a column dedicated to total deaths for each country. Let's start by visualizing that.

```python
# Plotting the total deaths of the top 50 most populated countries from covid-19 

death = top_50.plot(x="iso_code", y="total_deaths", kind="bar", figsize = (15, 15), title = "Total deaths from Covid-19") 
death.set_xlabel("Countries")
death.set_ylabel("total deaths")
```

Similarly, we can also visualize the column for total cases

```python
# Plotting the total cases of the top 50 most populated countries from covid-19 


cases = top_50.plot(x="iso_code", y="total_cases", kind="bar", figsize = (15, 15), title = "Total cases from Covid-19") 
cases.set_xlabel("Countries")
cases.set_ylabel("total cases")
```

While looking at total deaths and cases can give insight into how much a country has been affected, it doesn't take into account the population of that country. For example, Country A may have only 1000 cases, which is miniscule when compared to 1 million cases in Country B. However, Country A may have a population of 10000, while Country B has a population of a billion. In that case, Country A has a larger proportion of its population that has been impacted by COVID. 

Even though we are only looking at the top 50 most populous countries, this difference between totals and population is still important. The most populous in our top_50 list, China, has a 1.4 billion population, while the least populous, Venezuela is 28 million. We think this difference in population is large enough, so let's visualize proportion of deaths and cases next.

```python
# adds the proportion of total cases and total deaths compared to the population of the country to top_50
proportion_of_total_cases = []
proportion_of_total_deaths = []
for index, row in top_50.iterrows():
    proportion_of_total_cases.append(row['total_cases'] / row['population'])
    proportion_of_total_deaths.append(row['total_deaths'] / row['population'])
    
top_50['total_cases / population'] = proportion_of_total_cases
top_50['total_deaths / population'] = proportion_of_total_deaths
top_50
```

```python
# Plotting the cases / population of the top 50 most populated countries from covid-19 

cases_ratio = top_50.plot(x="iso_code", y="total_cases / population", kind="bar", figsize = (15, 15), title = "Covid-19 cases with respect to countries population") 
cases_ratio.set_xlabel("Countries")
cases_ratio.set_ylabel("total cases")
```

```python
# Plotting the deaths / population of the top 50 most populated countries from covid-19

death_ratio = top_50.plot(x="iso_code", y="total_deaths / population", kind="bar", figsize = (15, 15), title = "Covid-19 deaths with respect to countries population") 
death_ratio.set_xlabel("Countries")
death_ratio.set_ylabel("total deaths")
```

Our previous visualizations for total deaths/cases and proportion of deaths/cases provides insight on how the top 50 most populous countries have been impacted so far by the pandemic. 

In the data cleaning step, we added a column for the average amount of new deaths and new cases for each country. This average gives a clue on how much cases and deaths will increase in the future - how each country will be impacted. 

```python
avg_new_deaths_plot = top_50.plot(x="iso_code", y="avg_new_deaths", kind="bar", figsize = (15, 15), title = "Average new deaths for COVID-19") 
avg_new_deaths_plot.set_xlabel("Countries")
avg_new_deaths_plot.set_ylabel("Average amount of new deaths")
```

```python
avg_new_cases_plot = top_50.plot(x="iso_code", y="avg_new_cases", kind="bar", figsize = (15, 15), title = "Average new cases for COVID-19") 
avg_new_cases_plot.set_xlabel("Countries")
avg_new_cases_plot.set_ylabel("Average amount of new cases")
```
